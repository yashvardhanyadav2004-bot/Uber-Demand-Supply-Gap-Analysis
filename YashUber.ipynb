BUSINESS OBJECTIVES 1 Which ride status occurs most?

```python
df['Status'].value_counts()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x='Status',
    data=df,
    palette='Set2'
)

plt.title("Ride Status Distribution")

plt.xlabel("Ride Status")

plt.ylabel("Number of Requests")

plt.xticks(rotation=20)

plt.show()
```

Insight

Only 41.97% rides were successfully completed.

Failed rides (Cancelled + No Cars Available) completed rides se zyada hain.


BUSINESS OBJECTIVE 2
Demand by Time Slot

```python
df.groupby('Time_Slot')['Request id'].count()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x='Time_Slot',
    data=df,
    palette='viridis'
)

plt.title("Demand by Time Slot",fontweight='bold')

plt.xticks(rotation=45)

plt.show()
```

Insight

Morning time slot me sabse zyada requests aayi.

Early Morning me sabse kam demand thi.


BUSINESS OBJECTIVE 3
Status vs Time Slot

```python
pd.crosstab(df['Time_Slot'],df['Status'])
```

Heatmap

```python
plt.figure(figsize=(8,5))

sns.heatmap(
    pd.crosstab(df['Time_Slot'],df['Status']),
    annot=True,
    fmt='d',
    cmap='YlGnBu'
)

plt.title("Status by Time Slot")

plt.show()
```

Insight

Morning aur Evening me ride failures sabse zyada hain.

No Cars Available cases peak hours me increase hote hain.

BUSINESS OBJECTIVE 4

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x='Pickup point',
    hue='Status',
    data=df,
    palette='Set2'
)

plt.title("Pickup Point Analysis",fontweight='bold')

plt.show()
```

BUSINESS OBJECTIVE 5
Weekday Analysis

```python
weekday_order=['Sunday','Monday','Tuesday','Wednesday','Thursday','Friday','Saturday']

plt.figure(figsize=(8,5))

sns.countplot(
x='WeekDay',
data=df,
order=weekday_order,
palette='coolwarm'
)

plt.title("Weekday Demand Analysis")

plt.show()
```

Insight

Demand weekdays me relatively stable hai.

Koi extreme spike kisi ek weekday par nahi dikhta.

BUSINESS OBJECTIVE 6

 Demand vs Supply Gap

```python
status_counts=df['Status'].value_counts()

completed=status_counts['Trip Completed']

failed=status_counts['Cancelled']+status_counts['No Cars Available']

plt.figure(figsize=(6,6))

plt.pie(
[completed,failed],
labels=['Demand Met','Demand Unmet'],
autopct='%1.1f%%'
)

plt.title("Demand vs Supply Gap")

plt.show()
```

Insight

Large portion of ride demand fulfill nahi ho pa rahi.

Demand-supply gap business performance ko impact kar raha hai.
